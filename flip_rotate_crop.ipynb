```python
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("./assets/img.webp")
```

```python
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
```

```python
flip_horizontal=cv2.flip(image_rgb,1)
flip_vertical=cv2.flip(image_rgb,0)
flip_both=cv2.flip(image_rgb,-1)
```

```python
i,axs=plt.subplots(2,2,figsize=(8,6))

axs[0][0].imshow(image_rgb)
axs[0][0].set_title("Original")
axs[0][0].axis('off')

axs[0][1].imshow(flip_vertical)
axs[0][1].set_title("Vertical")
axs[0][1].axis('off')

axs[1][0].imshow(flip_horizontal)
axs[1][0].set_title("Horizontal")
axs[1][0].axis('off')

axs[1][1].imshow(flip_both)
axs[1][1].set_title("Both")
axs[1][1].axis('off')

plt.tight_layout()
plt.show()
```

```python
rotate_90_clock=cv2.rotate(image_rgb,cv2.ROTATE_90_CLOCKWISE)
rotate_90_anticlock=cv2.rotate(image_rgb,cv2.ROTATE_90_COUNTERCLOCKWISE)
rotate_180=cv2.rotate(image_rgb,cv2.ROTATE_180)

i,axs=plt.subplots(2,2,figsize=(12,6))


axs[0][0].imshow(image_rgb)
axs[0][0].set_title("Original")
axs[0][0].axis('off')

axs[0][1].imshow(rotate_90_clock)
axs[0][1].set_title("Rotate 90 Clock")
axs[0][1].axis('off')

axs[1][0].imshow(rotate_90_anticlock)
axs[1][0].set_title("Rotate 90 Anticlock")
axs[1][0].axis('off')

axs[1][1].imshow(rotate_180)
axs[1][1].set_title("Rotate 180")
axs[1][1].axis('off')

plt.tight_layout()
plt.show()
```
