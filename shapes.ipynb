```python
import cv2
import matplotlib.pylab as plt
import numpy as np
```

```python
height,widht=500,500
blue=(255,127,0)
canvas=np.full((height,widht,3),blue,dtype=np.uint8)
canvas_rgb=cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(canvas_rgb)
```

# Draw a line

```python
cv2.line(canvas_rgb,(50,50),(400,400),(255,255,255),5)
plt.imshow(canvas_rgb)
```

```python
cv2.rectangle(canvas_rgb,(10,10),(100,100),(125,125,125),5)
plt.imshow(canvas_rgb)
```

```python
cv2.circle(canvas_rgb,(100,100),50,(100,100,100),6)
plt.imshow(canvas_rgb)
```

```python
cv2.ellipse(canvas_rgb,(300,300),(100,50),45,0,360,(255,135,80),3)
plt.imshow(canvas_rgb)
```

```python
points=np.array([[150,400],[300,400],[225,300]],np.int32)
new_reshaped_points=points.reshape((-1,1,2))

print(new_reshaped_points)

cv2.polylines(canvas_rgb,[new_reshaped_points],False,(255,150,150),3)
plt.imshow(canvas_rgb)
```
